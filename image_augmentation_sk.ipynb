PIL (python imaging library) is pillow library which is used to read images


```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img=Image.open('/content/ironman.jpeg')
```

```python
def show(img,name):
  plt.imshow(img)
  plt.title(name)
  plt.show()
```

```python
show(img,'original ironman')
```

#Resize-transformation
moving from high resolution to low resolution



```python
import torchvision.transforms as transform
```

```python
resize_transformation = transform.Compose([transform.Resize(size=(100,100))])
```

```python
resized_img = resize_transformation(img)
```

```python
show(resized_img,"resized ironman")
```

# flipping transformation

```python
flipping_transformation = transform.Compose([transform.RandomHorizontalFlip(p=1)])
```

```python
flipped_img = flipping_transformation(img)
show(flipped_img,"flipped ironman")
```

```python
vertical_flipping_transformation = transform.Compose([transform.RandomVerticalFlip(p=1)])
```

```python
vertical_flipped_img = vertical_flipping_transformation(img)
show(vertical_flipped_img,"vertical flipped ironman")
```

#cropped image

```python
cropping_transformation = transform.Compose([transform.RandomCrop(size=(200,200))])
```

```python
cropped_img = cropping_transformation(img)
show(cropped_img,"cropped ironman")
```

#rotate transformation

```python
rotate_transformation = transform.Compose([[transform.RandomRotation(degrees=50)])
```

```python
rotated_img = rotate_transformation(img)
show(rotated_img,"rotated ironman")
```

# image normalization

```python
normalize_transformation=transform.Compose([
    transform.ToTensor(),
    transform.Normalize(mean=[0.5,0.5,0.5],std=[0.1,0.1,0.1]),
    transform.ToPILImage()
    ])
```

```python
normalized_img=normalize_transformation(img)
show(normalized_img,"normalized ironman")
```

# combining all augmentation together

```python
transformation=transform.Compose([
    transform.RandomHorizontalFlip(p=1),
    transform.RandomCrop(size=(200,200)),
    transform.RandomRotation(degrees=50)
    ])
```

```python
transformed_img = transformation(img)
show(transformed_img,"fully transformed ironman")
```
