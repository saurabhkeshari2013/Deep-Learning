```python
from gensim.models import Word2Vec
```

```python
sentences = ["harry went to hogwards",
             "hogwards is a school of wicthcraft",
             "hermoine was the topper of the class",
             "lucious malfoy was involved with voldemort",
             "the wand chooses the wizard",
             "dobby has no master",
             "voldermort himself created his worst enemy"
             ]
```

# convert to a token

```python
sent_tokens=[]
for sent in sentences:
  sent_tokens.append(sent.split(" "))
```

```python
sent_tokens
```

```python
model=Word2Vec(sent_tokens, min_count=1, vector_size=5)
```

```python
# summarise the vocabulary
words=list(model.wv.index_to_key)
words
```

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots()
for word in words:
  vec=model.wv[word]
  ax.scatter(vec[0],vec[1])
  ax.annotate(word,(vec[0],vec[1]))
```
